La Data Science est un domaine multidisciplinaire qui combine plusieurs compétences pour extraire des connaissances à partir des données. Voici les notions clés à maîtriser :  



## I. Manipulation des données  
Python et bibliothèques : Pandas, NumPy  
Nettoyage des données : Gestion des valeurs manquantes, détection des doublons  
Exploration des données (EDA) : Statistiques descriptives, visualisations  

Exemple en Python :  
```python
import pandas as pd
df = pd.read_csv("data.csv")
print(df.info())  # Aperçu des données
print(df.describe())  # Statistiques basiques
```

 ## II. Statistiques et probabilités 
Statistiques descriptives : Moyenne, médiane, écart-type  
Statistiques inférentielles : Test d'hypothèse, intervalles de confiance  
Probabilités : Loi normale, distribution binomiale, Bayes  

Exemple en Python :  
```python
import numpy as np
data = np.random.normal(loc=0, scale=1, size=1000)  # Génère des données selon une loi normale
```


## III. Visualisation des données 
Matplotlib & Seaborn pour graphiques  
Histogrammes, boîtes à moustaches, heatmaps  

Exemple en Python :  
```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["colonne"], kde=True)
plt.show()
```

## Machine Learning (Apprentissage automatique) 
Supervisé : Régression (linéaire, logistique), Classification (SVM, Random Forest, XGBoost)  
Non supervisé : Clustering (K-Means, DBSCAN), Réduction de dimension (PCA)  

Exemple en Python (régression linéaire) :  
```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

## Deep Learning (Apprentissage profond)
Réseaux de neurones : Perceptron, CNN (Vision), RNN (Séries temporelles)  
Frameworks : TensorFlow, PyTorch  

Exemple avec TensorFlow :  
```python
import tensorflow as tf
model = tf.keras.Sequential([tf.keras.layers.Dense(10, activation='relu')])
```


## Big Data et Cloud Computing  
Technologies : Hadoop, Spark  
Bases de données : SQL, NoSQL (MongoDB)  
Cloud : AWS, Google Cloud, Azure  

Exemple en SQL:  
```sql
SELECT AVG(salaire) FROM employes WHERE poste = 'Data Scientist';
```

## Ingénierie des fonctionnalités (Feature Engineering) 
Sélection des meilleures variables  
Encodage des variables catégorielles  
Normalisation et standardisation  


## Traitement du langage naturel (NLP)**  
Tokenization, Stemming, Lemmatisation  
Word Embeddings : Word2Vec, BERT  


## Éthique et biais en Data Science  
Protection des données (RGPD)  
 Équité et biais dans les modèles  

## Conclusion 
La Data Science est un domaine vaste qui combine statistiques, programmation et machine learning. Pour exceller, il faut pratiquer sur des projets réels!

````python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D,Flatten,Dense
from tensorflow.keras.utils import to_categorical
````

````python
#Selection des caractéristiques de la cible
features= ['Pclass','Sex','Age','SibSp','Parch','Fare']
target="Survived"
X=data[features]
y=data[target]
````

````python
#Normalisation des données
scaler=StandardScaler()
X_scaled= scaler.fit_transform(X)
````

````python
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,y, test_size=0.2, random_state=42)
````

````python
#Creationdu model DNN avec plusieurs couches
model=Sequential([Dense(128,activation='relu', input_shape=(X_train.shape[1],)),
Dense(64,activation='relu'),
Dense(32,activation='relu'),
Dense(1,activation='softmax')
])
````

````python
#complilation du model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
````

````python
#Entrainement du model
model.fit(X_train,Y_train,epochs=50,batch_size=32,validation_split=0.2, verbose=0)
````

````python
#Prediction sur l'ensemble des données test
y_pred=model.predict(X_test)
````

````python
y_pred_binary=(y_pred>0.5).astype(int)
y_pred
y_pred_binary
````

````python
#Évaluation du model/ calcule des metrics
precision=accuracy_score(Y_test,y_pred_binary)
print(f'Precision du modele est: {round(precision*100,0)}'"%")
````
